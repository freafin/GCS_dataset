```python
import numpy as np
import os
import cv2  
```

```python
def compute_global_std(image_dir):
    pixel_values = []
    for fname in os.listdir(image_dir):
        if fname.endswith(".jpg"):  # or .jpg, .tif
            img = cv2.imread(os.path.join(image_dir, fname), cv2.IMREAD_GRAYSCALE)
            pixel_values.append(img.flatten())
    all_pixels = np.concatenate(pixel_values)
    return np.std(all_pixels)

std_gcs = compute_global_std("path/to/GCS/images")
print(f"GCS global pixel intensity std: {std_gcs:.2f}")
```
