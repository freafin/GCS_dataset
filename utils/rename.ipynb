```python
import os

folder_path = "insert_path"

def rename_files_in_folder(folder_path):
    """
    Renames all files in a folder by adding an underscore (_) after the nth character.
    :param folder_path: Path to the folder containing files to rename.
    """
    for filename in os.listdir(folder_path):
        old_path = os.path.join(folder_path, filename)

        # Ensure it's a file (not a directory)
        if os.path.isfile(old_path):
            # Insert underscore after the nth character
            new_filename = filename[:2] + "_" + filename[2:]

            new_path = os.path.join(folder_path, new_filename)
            os.rename(old_path, new_path)

            print(f"Renamed: {filename} â†’ {new_filename}")


rename_files_in_folder(folder_path)
```
