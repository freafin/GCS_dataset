```python
import os
import json

def count_sam2_instances(json_dir):
    print(f"{'File':<20} {'Instances':>10}")
    print("-" * 32)
    total = 0
    for fname in sorted(os.listdir(json_dir)):
        if fname.endswith(".json"):
            path = os.path.join(json_dir, fname)
            try:
                with open(path, "r") as f:
                    data = json.load(f)
                    count = len(data.get("annotations", []))
                    print(f"{fname:<20} {count:>10}")
                    total += count
            except Exception as e:
                print(f"{fname:<20} [Error: {e}]")
    print("-" * 32)
    print(f"{'Total':<20} {total:>10}")

# Example usage
count_sam2_instances("insert_path")
```
