```python
import os

def count_instances_in_labels(label_dir):
    print(f"{'File':<20} {'Instances':>10}")
    print("-" * 32)
    total = 0
    for fname in sorted(os.listdir(label_dir)):
        if fname.endswith(".txt"):
            path = os.path.join(label_dir, fname)
            with open(path, "r") as f:
                count = sum(1 for line in f if line.strip())
                print(f"{fname:<20} {count:>10}")
                total += count
    print("-" * 32)
    print(f"{'Total':<20} {total:>10}")

# Example usage
count_instances_in_labels("insert_path")
```
