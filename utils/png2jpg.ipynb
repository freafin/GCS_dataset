```python
from PIL import Image
import os

input_dir = "insert_path"
output_dir = "insert_path"

for filename in os.listdir(input_dir):
    if filename.endswith(".png"):
        img_path = os.path.join(input_dir, filename)
        img = Image.open(img_path).convert("RGB")  # ensure no alpha channel
        new_filename = filename.replace(".png", ".jpg")
        new_path = os.path.join(output_dir, new_filename)
        img.save(new_path, "JPEG", quality=100)
        print(f"Converted: {filename} â†’ {new_filename}")
```
