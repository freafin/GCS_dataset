```python
import os
from ultralytics import YOLO
from matplotlib import pyplot as plt
from PIL import Image
```

```python
#Instansiate
model = YOLO('yolo11n-seg.yaml')  # build a new model from YAML
model = YOLO('TNnano_seg_transfer_best.pt')  # Transfer the weights from a pretrained model (recommended for training)
```

```python
# define number of classes based on YAML
import yaml
with open("data.yaml", 'r') as stream:
    num_classes = str(yaml.safe_load(stream)['nc'])
```

```python
#Define a project --> Destination directory for all results
project = "C:/Users/fre_f/PythonProjects/CellVision/data/GC/GC_validated/yolo/unpatched/results"
#Define subdirectory for this specific training
epochs = 1000
name = f"transfer_unpatched_yolo11nano_{epochs}_epochs-" #note that if you run the training again, it creates a directory: 200_epochs-2
```

## Train

```python
# Train the model
results = model.train(data='data.yaml',
                      project=project,
                      name=name,
                      epochs=epochs,
                      patience=0,
                      optimizer='AdamW',
                      lr0=0.001,
                      batch=1,
                      device=0,
                      verbose=False,
                      imgsz=512,
                      val=False)
```
