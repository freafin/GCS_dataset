```python
from PIL import Image 
import os 
```

```python
image_path = "insert_path"
for img in os.listdir(image_path):
    if img.lower().endswith('.jpg'):
        full_img_path = os.path.join(image_path, img)
        input_image = Image.open(full_img_path).convert('RGB')

        # Replace '.jpg' with '.png'
        base_name = os.path.splitext(img)[0]
        new_img_path = os.path.join(image_path, base_name + '.png')

        input_image.save(new_img_path, 'PNG')
        os.remove(full_img_path)  # Optionally remove the original JPG
```
